# **Histogram**

- Variable : Continous
- type of plot : Univariate
- Importance : Used to check the distribution of data,x axis contains a variable and y axis will be a frequency of that variable

```python
plt.hist(df["Grade"])

# Here to consider bin size
# To calculate bin size
# Bin Size = (Max - Min)/No of Bins
# Here the number of bins in constant --> 10
# Bin Size = (165 - 3)/10  == 16.2
# In the result we can see the array which increases by 16.2
```

- **This customization option is not available in SEABORN**

```python
# If you want to customize the number of bins
plt.hist(df["Grade"], bins = 10, range = (0,200))
# Bins size = (200 - 0)/10 == 20
```

- **Here we can see that x = 25 is not aligned with the graph shape for that we use the followingly**

```python
plt.hist(df["Grade"], bins = 10, range = (0,200))
plt.xticks(list(range(0,201,20)))
```

```python
plt.hist(df["Grade"])
plt.show()
```

```python
plt.hist(df["Grade"])
plt.xlabel("Grade")
plt.ylabel("No of Students")
plt.show()
```

```python
plt.boxplot(df["Grade"])
plt.show()
```

- **To Extract Outlier**

```python
Q3 = df["Grade"].quantile(0.75)
Q1 = df["Grade"].quantile(0.25)
IQR = Q3- Q1
upper_limit = Q3+(1.5*IQR)
lower_limit = Q1-(1.5*IQR)

df[(df["Grade"]>upper_limit) | (df["Grade"]<lower_limit)]
```

- **Stacked Barplot**

```python
turnover = [90,136,89,27]
profit = [40,16,34,12]
companies= ['GOOGLE', 'AMZN', 'MSFT', 'FB' ]

plt.bar(companies,turnover, label="turnover")
plt.bar(companies,profit, label="profit")
plt.legend()
plt.show()
```

- Remember that we should apply this only when we have first plot value higher than the second plot

- **Subplots**

```python
#Program for drawing multiple bar charts on one image.
ecommerce=['Snapdeal', 'Alibaba', 'Amazon', 'Flipkart']
Q1_Profit=[45, 100, 70, 40]
Q2_Profit=[40, 105, 65, 45]
Q3_Profit=[42, 120, 72, 50]
Q4_Profit=[34, 115, 60, 69]

#Creating different bar charts on one image using subplot () function.
plt.figure(figsize=(10,12))

#Creating bar chart in first cell of figure having 2 rows, 2 columns.
plt.subplot(2,2,1)
plt.bar(ecommerce,Q1_Profit)

#Creating a bar chart in second cell.
plt.subplot(2,2,2)
plt.bar(ecommerce,Q2_Profit)

#Creating a bar chart in third cell.
plt.subplot(2,2,3)
plt.bar(ecommerce, Q3_Profit)

#Creating a bar chart in fourth cell.
plt.subplot(2,2,4)
plt.bar(ecommerce,Q4_Profit)

#Displaying the chart
plt. show ()
```
