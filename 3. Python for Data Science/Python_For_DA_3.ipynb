# **Using Multiple conditions**

- **Syntax:**   df[ (condition-1) or/and (condition-2)]

```python
df[(df["tip"] >=7 ) & (df["gender"] == "Male" ) ]                          # And Condition
```

```python
df[(df["tip"] >=7 ) | (df["gender"] == "Male" ) ]                # OR condition
```

```python
df[(df["tip"] >=3) | (df["gender"] == "Female") & (df["total_bill"] <= 60) ]
```
