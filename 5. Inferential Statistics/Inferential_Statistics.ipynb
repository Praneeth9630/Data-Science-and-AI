**Case Study 1 :** The length of 25 samples of a fabric are taken at random. Historic Mean and standard deviation are 150 cm and 5.5 cm respectively. Company Manager wants to test if the current mean is equal to the the historic mean. Assume a to be 0.05

**Step 1. Formulation of Ho,Ha**

- Ho: Mean = 150 cm
- Ha: Mean != 150 cm

**Step 2. Select Level of significance**

- alpha = 0.05

**Step 3. Check for Normality**

```python
import numpy as np
import pandas as pd

from scipy import stats
```

```python
df1 = pd.read_excel("fabric data.xlsx")
df1.head()
```

```python
df1['Fabric_length']. skew()  # As the skewness is between -1 to +1 it is a normal distribution
```

**Step 4. Select the statistical test and check p value**

- 1 sample Ztest with two tail rejection

```python
from statsmodels.stats.weightstats import ztest
```

```python
ztest(df1["Fabric_length"],value=150, alternative = 'two-sided', ddof=0)
```

```python
df1["Fabric_length"].mean()
```

**Step 5. Based on p value, Accept or Reject H0**

- P < alpha
- P Low -- Null go
- Reject H0 -- > Mean != 150 (action required)

specification value.

**1. Formulation of H0,H1**

## **ANOVA Test**

**Case Study 3 : Creditcard Promotion**

A financial analyst at a Financial institute wants to evaluate a recent credit card promotion. After this promotion, 500 cardholders were randomly selected. Half
received an ad promoting a full waiver of interest rate on purchases made over the next three months, and half received a standard Christmas advertisement.
Did the ad promoting full interest rate waiver, increase purchases?

**1. Formulation of H0,H1**

- Ho (No difference) : Avg of purchases made by FIW =Avg purchases made by SC =>default/ current/ no action
- Ha : Avg of purchases made by FIW != Avg purchases made by SC =>take action

**2. Select Level of significance**

- alpha = 0.05

**3. Check for Normality**

```python
df3=pd.read_excel("Promotion.xlsx")
df3.head()
```

```python
df3['InterestRateWaiver'].skew(), df3['StandardPromotion'].skew()
```

**4. Select the Statistical test & Calculate the p value**

- ANOVA Test - Analysis Of Varaiance(F-Test)

```python
anova_test = stats.f_oneway(df3['InterestRateWaiver'], df3['StandardPromotion' ])
anova_test
```

**5. Based on p value, Accept or Reject H0**

- P < alpha
- P Low -- Null go
- Reject H0 (FIW=SC) -- > take action

# **Chi-Square Test**

**Case Study 5 :** Johnnie Talkers soft drinks division sales manager has been planning to launch a new sales incentive program for their sales executives. The sales executives felt that adults (>40 yrs) won't buy, children will & hence requested sales manager not to launch the program. Analyze the data & determine whether there is evidence at 5% significance level to support the hypothesis

**1. Formulation of H0,H1**

- H0 -> Proportions of Adults = Proportions of Children
- H1 -> Proportions of Adults != Proportions of Children

**2. Select Level of significance**

- alpha = 0.05

**4. Select the Statistical test & Calculate the p value**

- Chi-Square Test

```python
df5=pd.read_excel("JohnyTalkers.xlsx")
df5.head()
```

```python
df5.info()
```

```python
df5["Person"].value_counts()
```

```python
df5["Drinks"].value_counts()
```

```python
count = pd.crosstab(df5["Person"],df5["Drinks"],margins=True)
count
```

```python
chisquare_test = stats.chi2_contingency(count)
chisquare_test
```

**5. Based on p value, Accept or Reject H0**

- P < alpha
- P low Null go
- Reject H0
- Accept alternate hypothesis i.e. Unequal proportions
