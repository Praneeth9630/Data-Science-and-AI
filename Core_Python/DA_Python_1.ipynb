# Rules in Naming Variables
- Rule 1: A variable name must start with a letter (a-z, A-Z) or the underscore character
- Rule 2: A variable accepts only underscore_symbol only (No other symbol is accepted)
- Rule 3: A variable name cannot start with a number
- Rule 4: A variable name can contain combination of alpha-numeric characters and underscores
- Rule 5: Variable names are Case Sensitive (age, Age and AGE are three different variables)
- Rule 6: It can't accept reserve/keywords
- List of Keywords: ['False', 'None', 'True', 'and', 'as', 'assert', 'async', 'await', 'break', 'class', 'continue', 'def', 'del', 'elif', 'else', 'except', 'finally', 'for', 'from', 'global', 'if', 'import', 'in', 'is', 'lambda', 'nonlocal', 'not', 'or', 'pass', 'raise', 'return', 'try', 'while', 'with', 'yield']
