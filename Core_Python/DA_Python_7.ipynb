# **Set Methods**

**set.add()**
- add method adds a element to the set. if the element already exists in the set nothing will happen


```python
s = {1,2,3}
s.add(4)
print(s)
```

```python
s = {10,20,30}
s.add(40)
s.add(50)
s.add("surya")
print(s)
```

# **set.update()**
- **update** method updates the set with another set or any iterable

```python
s = {10,20,30}
s.update([40,50,"surya"])
print(s)
```

```python
s = {1,2,3}
s.update([4,5,"abc",True,(6,7,8)])   #Lets add a tuple in the set
print(s)
```

# **set.discard()**
- **discard** method removes the specified item from the set. if the item doesnot exist in the set, nothing happens

```python
s = {1,2,3}
s.discard(2)
print(s)
```

```python
s = {1,2,3}
s.discard(5)
print(s)
```

# **set.remove()**
- **remove** method removes the specified element from the set. if the element doesnt exist in the set, a key error is raised

```python
s = {1,2,3}
s.remove(2)
print(s)
```

```python
s = {1,2,3}
s.remove(5)
print(s)
```

# **set.clear()**
- **clear** method removes all elements from the set

```python
s = {1,2,3}
s.clear()          # removes all items in the set
print(s)           # Remember in set the empty set is represented by set() not {}
```

```python
s = {1,2,3}
del s              # delets the variable
print(s)
```

# **set.copy()**
- **copy** method returns a shallow copy of the set
- changes to the copied set doesnt effect the original set

```python
s1 = {1,2,3}
s2 = s1.copy()
s2.add(4)
print(s1)
print(s2)
```

# **set.union()**
- This method returns a new set with elements from both sides

```python
setA = {1,2,3}
setB = {2,3,4}
setA.union(setB)
```

# **set.intersection()**
- **intersection** method returns a new set with elements that are common to all sets

```python
setA = {1,2,3}
setB = {2,3,4}
setA.intersection(setB)
```

# **set.isdisjoint()**
- **isdisjoint** method returns True if two sets have no intersection, otherwise it will return false

```python
s1 = {1,2,3}
s2 = {4,5,6}
s3 = {4,7,8}

s1.isdisjoint(s2)
```

```python
s2.isdisjoint(s3)
```

# **set.difference()**
- **difference** method returns a new set with elements in the set that are not in the other specified sets

```python
set1 = {1,2,3}
set2 = {2,3,4}

print(set1 - set2)     # within set1 remove the set2 items
print(set2 - set1)     # within set2 remove the set1 items
```

# **set.symmetric_difference()**
- **symmetric_difference** method returns a new set with elements that are either of the sets, but not in both
- (A union B) - (A intersection B)


```python
setA = {1,2,3,"praneeth"}
setB = {2,3,4}

setA.symmetric_difference(setB)           #with union remove intersection items
```

```python
setA = {1,2,3,"praneeth"}
setB = {2,3,4}

setA.union(setB)  -  setA.intersection(setB)
```

# **set.issubset()**
- **issubset** method returns True if all elements of first set are available in second set, otherwise it returns false

```python
a = {2,3,4}
b = {2,3,4,5}

ans = a.issubset(b)
print(ans)
ans = b.issubset(a)
print(ans)
```

# **set.issuperset()**
- **issuperset()** method returns True if all elements of second set are available in first set, otherwise it returns false

```python
a = {2,3,4}
b = {2,3,4,5}

ans = a.issuperset(b)
print(ans)
ans = b.issuperset(a)
print(ans)
```
