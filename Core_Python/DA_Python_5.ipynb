# **List**
- It is one of the most used datatype in Python and is very flexible.
- List is Advance Data type, where it can store multiple values
- Each value inside the list is called an item.
- A list can have from 0 items to multiple items
- Declaring a list is, Items separated by commas are enclosed within brackets [].
- All the items in a list do not need to be of the same type (heterogenous).
- List is an ordered sequence of items.
- items can be duplicated (list items can be repeated)

# **Lists are Mutable**
- Mutable: value of elements of a list can be altered in the same object
- Modification: add, remove, replace

```python
a = "siva"
a
```

```python
a = "praneeth"
a
```

String is immutable but above example is possible because it creates new object "a" by replacing old "a"

```python
a = "siva"
id(a)
```

```python
a = a.replace("siva","praneeth")
print(a)
id(a)
```

# **list.sort()**
- used to sort items in a list either in ascending order or descending order
- by default, its sort in ascending order
- **Note** Sort is applicable for either only alphabet or only numeric values

```python
l = [22,11,99,44,33,1,10]
l.sort()                                          #ascending order
print(l)
```

```python
l = [22,11,99,44,33,1,10]
l.sort()
l.reverse()                              #descending order
print(l)
```

```python
l = [22,11,99,44,33,1,10]
l.sort(reverse=True)                              #descending order
print(l)
```

```python
l = ["surya","praneeth","kishore","abhi"]
l.sort()
print(l)
```

```python
l = ["surya","praneeth",1,2,3]
l.sort()
print(l)
```

```python
l = [1,2,3,[4,5],6]
l.sort()
print(l)
```

# **list.copy()**
- used to create new list object with a copy of items

```python
a = [1,2,3,4,5]
b = a.copy()
b.append(10)
print(b)
print(a)
```

**Difference between shallow copy and deep copy**

![Screenshot 2024-07-08 120928.png](
  
)

```python
# Shallow copy  --- It indicates to different objects
a = [1,2,3,4,5]
b = a.copy()
b.append(10)
print(b)
print(a)
```

```python
# Deep copy  --- It indicates to same objects
a = [1,2,3,4,5]
b = a
b.append(10)
print(b)
print(a)
```
