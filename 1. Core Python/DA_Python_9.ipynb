```python
x = 5
print(x)
```

```python
x = 10
x = x + 5
print(x)
```

```python
x = 10
x += 5
print(x)
```

# **Logical Operators / Boolean Operators**
- Logical Operators are used to check multiple conditions at a time.
- It returns bool type only
- and, or, not are Logical operators.
- when we have, "or" operator...atleast 1 condition should be True then only overall output will be True
- when we have, "and" operator...all conditions should be True then only overall output will be True

**Logical operators Precedence**

- Logical OR

# **Identity opertors**
- Identity @perators are used to check if two values (or variables) are indicating to same object or not
- It returns output as boolean
- is, is not are the identity operators in Python.
- is operator (# is - True if the opernds are identical)
- is not operator (# is not - True if the operands are not identical)

```python
a = [1,2,3,4]
b = a

a is b
```

```python
id(a)
```

```python
id(b)
```

```python
a = [1,2,3,4]
b= a.copy()

a is b
```

```python
id(a)
```

```python
id(b)
```

```python
a = 5
b = 5

print(a is b)         # id(a) == id(b)
```

```python
s1 = "satish"
s2 = "satish"

print(s1 is s2)      #id(s1) == id(s2)
```

```python
a = 6
b = 8

a is not b
```

# **Membership Operators**
- Membership operators are used to test whether a value or variable is found in a sequence (string, list, tuple, set and dictionary).
- It returns oputput as boolean
- in, not in are the membership operators.

```python
a = "venkatesh"

"esh" in a
```

```python
l = [1,2,3,4]

5 in l
```

```python
a = [1,2,3,4,5]

9 not in a
```
