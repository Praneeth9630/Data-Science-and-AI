# **Tuples**
1. Tuple is similar to List except tuple are immutable which means we cannot change the elements of a tuple once assigned.
2. When we do not want to change the data over time, tuple is a preferred data type.
3. Iterating over the elements of a tuple is faster compared to iterating over a list.

**Empty tuple**

```python
t = []
type(t)
```

```python
tup = ()
type(tup)
```

```python
tup = ()
print(tup)
```

```python
t = tuple()
print(t)
```

**Tuple is Heterogeneous**

```python
t = (1,0.5,True,"praneeth",[1,2,3])
print(t)
```

**Nested tuple**

```python
t = (1,0.5,True,"praneeth",[1,2,3],(10,20,30), 100)
print(t)
```

**Tuple indexing**

```python
t = (1,0.5,True,"praneeth",[1,2,3],[(10,20,30), 100])
t[5]
```

```python
t[5][0]
```

```python
t[5][0][0]
```

```python
t[5][1]
```

**Tuple** **Slicing**

```python
t = (1,2,3,4)
t[:3]
```

```python
t = (1, [2,3,4], 5, [6,7,8], 9)
t[1:3]
```

```python
t[1][:3]
```

**Tuple concatenation**

```python
t1 = (1,2,3)
t2 = (4,5,6)
t1 + t2
```

# **Tuple methods**

**tuple.count()**
- No of times item occurred in tuple

```python
 t = (1,2,3, 1,2,3, [1,2,3])
 t.count(1)
```

**tuple.index()**
- index number of first occurence of item

```python
 t = (1,2,3, 1,2,3, [1,2,3])
 t.index(1)
```

**can't replace a item in the tuple**

```python
t = (1,2,3,4)
t[1] = 40
print(t)
```

```python
t = (1,2,3,4)
del t[1]
print(t)
```
