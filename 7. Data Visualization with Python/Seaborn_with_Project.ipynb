# **Scatter Plot**

```python
sns.scatterplot(x=df['total_bill'],y=df['tip'])
plt.show()
```

**scatter plot using 3 variables**

- 2 Continous Variables
- 1 Discrete Variable

Basically scatter plot is applied on 2 continous variables, if you want 3rd variable, it should be compulsorly discrete

```python
lunch = df[df["time"] == "Lunch"]

dinner = df[df["time"] == "Dinner"]
print(dinner)

plt.scatter(x=lunch["total_bill"],y=lunch["tip"],color="blue", label = "Lunch")
plt.scatter(x=dinner["total_bill"],y=dinner["tip"],color = "green", label = "Dinner")
plt.legend()
plt.show()
```

```python
sns.scatterplot(x=df['total_bill'],y=df['tip'],hue=df["day"])
plt.show()
```

```python
sns.relplot(x=df['total_bill'],y=df['tip'],hue=df["day"])      # Diff btw scatterplot and replot is legend will be outside in replot
plt.show()
```

```python
sns.relplot(x=df['total_bill'],y=df['tip'],col=df["day"])
plt.show()
```

```python
sns.relplot(x=df['total_bill'],y=df['tip'],col=df["day"],col_wrap=2)
plt.show()
```

```python
lunch = df[df["time"] == "Lunch"]

dinner = df[df["time"] == "Dinner"]

plt.subplot(1,2,1)
plt.scatter(x=lunch["total_bill"],y=lunch["tip"])
plt.subplot(1,2,2)
plt.scatter(x=dinner["total_bill"],y=dinner["tip"])

plt.show()
```

- We use **Col** if we want **Multiple plot or each plot for each category**
- We use **hue** if we want **Single plot with Different colors**

# **JointPlot**

A join plot allows to study the relationship between 2 numeric variables. The central chart display their correlation. It is usually a scatterplot, a hexbin plot, a 2D
histogram or a 2D density plot

```python
sns.jointplot(x='total_bill',y='tip',data=df)
plt.show()
```

# **Violin Plot**

Violin plot helps us to see both the distribution of data in terms of Kernel density estimation and the box plot

```python
sns.violinplot(x="total_bill", y="day",data=df)
plt.show()
```

# **Pair plot - Multiple continous variables**
A "pairs plot" is also known as a scatterplot, in which one variable in the same data row is matched with another variable's value, like this: Pairs plots are just
elaborations on this, showing all variables paired with all the other variables

```python
sns.pairplot(df, vars=continous)
plt.show()
```

```python
sns.pairplot(df,vars=["total_bill","tip"])
plt.show()
```

- **When you have 2 variables then join plot is the best and when you have moe variables pair plot is the best**
